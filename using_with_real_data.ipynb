С помощью Etherscan получим список последних транзакций Etherium с одного кошелька на другой в виде csv-файла.

```python
from selenium import webdriver
from os import listdir
opt = webdriver.ChromeOptions()
opt.pageLoadStrategy = "eager"
opt.add_experimental_option("prefs", {
  "download.default_directory": download_path,
  "download.directory_upgrade": True,
    "download.enable_downloads": True,
})

driver = webdriver.Chrome(options=opt)

driver.get("https://etherscan.io/txs")

driver.find_element(webdriver.common.by.By.ID, "btnExportQuickTransactionListCSV").click()

table_path = download_path + '\\' + list(filter(lambda x : x.startswith("export-transaction-list-"), listdir(download_path)))[0]
table_path
```
